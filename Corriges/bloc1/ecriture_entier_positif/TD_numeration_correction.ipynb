# Exercice 6

Q1. Rappeler le nombre de valeurs possibles que peut prendre un nombre $N$ écrit sur $n$ bits .\
Faire l'application numérique pour un nombre $N$ écrit sur 1 octet, puis pour un nombre $N$ écrit sur 2 octets.

**Réponse**

nombre de valeurs possibles d'un nombre écrit sur n bits = $2^n$  
nombre de valeurs possibles d'un nombre écrit sur 1 octet = $2^8 = 256$  
nombre de valeurs possibles d'un nombre écrit sur 2 octet = $2^{16} = 65536$

Q2. Donner la valeur  maximale d'un nombre $N$  écrit sur $n$ bits .\
Faire l'application numérique pour un nombre $N$ écrit sur 10 bits, puis pour un nombre $N$ écrit sur 2 octets.

**Réponse**

$N_{MAX}=2^n-1$  
* sur 10 bits $N_{MAX}=2^{10}-1 = 1023$  
* sur 2 octets $N_{MAX}=2^{16}-1 = 65535$

Q3. Quel est le nombre minimum de bits $n$  nécessaires pour écrire un nombre $N$ prenant des valeurs comprises entre 0 et 99 ? 
*(Pour cela, écrire une équation, mais ne pas chercher à la résoudre. Procéder plutôt par essais...)*

**Réponse**

* avec n = 6 bits : $N_{MAX}=2^6-1 = 63$  
* avec n = 7 bits : $N_{MAX}=2^7-1 = 127$

$\Longrightarrow$ il faut donc au moins 7 bits pour coder des valeurs comprises entre 0 et 99

Q4. Même question que précédemment mais pour un nombre $N$ prenant des valeurs comprises entre 0 et 12756.

**Réponse**

* avec n = 13 bits : $N_{MAX}=2^{13}-1 = 8191$  
* avec n = 14 bits : $N_{MAX}=2^{14}-1 = 16383$

$\Longrightarrow$ il faut donc au moins 14 bits pour coder des valeurs comprises entre 0 et 12756


# Exercice 7

> Dans un réseau, les ordinateurs fonctionnant avec le protocole TCP/IP sont identifiés par une adresse. En IPv4, cette adresse est composée de quatre nombres. Par exemple, une machine pourrait avoir l’adresse 172.16.0.3. Ceci est une adresse IP. Chaque nombre séparé par un point est codé sur un octet. **L’adresse est donc constituée de 4 octets**.

1. Sur combien de bits est codé chaque nombre d’une adresse IP ? Sachant cela, quelle est la taille en bits d’une adresse complète ?

**Réponse**

* Chaque nombre d'une adresse IP est codé sur 1 octets donc sur 8 bits
* Une adresse IP a donc une taille de $4 \times 8 = 32$ bits


2. Pour chaque nombre, donnez l’intervalle de valeurs possibles. En déduire le nombre d’adresses différentes que l’on peut théoriquement former.

**Réponse**

* Chaque nombre d'une adresse IP peut prendre des valeurs comprises entre $0$ et $2^8-1=255$ soit $256$ valeurs possibles
* Il y a théoriquement donc $256^4$ adresses IPv4 possibles (soit plus de 4 milliards)


> Les adresses IP sont structurées :
* une partie identifie le réseau
* une partie identifie l’appareil dans le réseau.

> Prenons un exemple : Imaginons que l’on est sur la machine d’adresse IP : 172.16.1.80. À cette adresse est associé un masque 255.255.255.0. La notation binaire indique que les 24 premiers bits sont à 1, et les 8 derniers sont à 0. Les bits à 1 permettent d'identifier le réseau et les bits à 0 permettent d'identifier la machine sur le réseau.
* 172.16.1  est donc la partie **réseau** de l'adresse IP
* 80 est donc la partie **machine** de l'adresse IP

> Pour noter un masque et une adresse IP de manière plus synthétique, on utilise cette notation : « @IP / n » où n est le nombre de bit à 1 du masque. Ici, on aurait 172.16.1.80 / 24. [Plus d'informations ici](https://fr.wikipedia.org/wiki/Adresse_IP)

On considère une machine dont l’adresse IP est 192.168.3.1/27. 

1. Combien de bits sont disponibles pour identifier les machines sur le réseau ?

**Réponse**

27 bits

2. Écrire en binaire le masque, puis le convertir en décimal.

**Réponse**

Le masque possède 27 bits à 1 et donc 5 bits à 0
masque = 11111111 11111111 11111111 11100000


3. Combien de machines peut-on théoriquement connecter sur ce réseau ?

**Réponse**

Il y a 5 bits servant à identifier la machine, ce qui laisse $2^5 = 32$ possibilités. On peut donc connecter au plus 32 machines sur ce réseau
