## Synthèse intermédiaire

### Programmation des événements

* _on**event**_ est un **attribut**  permettant de spécifier l'événement qui nous intéresse. Comme tout attribut HTML, _on**event**_ est noté dans la balise ouvrannte de la **cible**. Par exemple `onclick`, `onmouseover` etc...


* la valeur de cet attribut correspond au code javascript qui sera exécuté lorsque l'événement apparaît. Il est plus lisible de mettre comme code javascript un simple **appel de fonction**


* Le corps de la fonction est écrit entre des balises `<script> </script>`. Le tout placé juste avant la balise `</body>` (comme dans l'activité 2)


* Il est encore plus clair d'écrire le code javascript dans un fichier séparé d'extension **.js**. Dans ce cas, on mettra juste avant la balise `</body>` le _lien_ vers le fichier javascript

### Quelques caractéristiques du langage javascript

> Javascipt possède beaucoup de similitudes avec python et quelques différences. L'apprentissage de javascript demande beaucoup de temps, que nous n'avons pas en NSI : On n'attend pas de vous la même maîtrise en javascript et en python ! 

* javascript est un langage **orienté objet** utilisant la notation pointée (comme python). l'objet `window` représente la fenêtre du navigateur. `window.alert` permet d'appeler une méthode (ou fonction) `alert` qui affiche une nouvelle fenêtre "popup" dans la fenêtre `window`. La fonction `alert` prend en paramètre une chaîne de caractères correspondant au texte affiché dans la fenêtre "popup"


* chaque instruction doit être suivie d'un point-virgule : ;


* ce n'est pas l'indentation qui structure le code : les instructions d'un même bloc de code sont ici encadrées par des accolades : { .... }


* les commentaires sont précédées d'un double-slash : //

## Synthèse intermédiaire

* L'objet `document` représente le code HTML de la page web (plus précisément son arbre DOM)


* La méthode `getElementById` permet de **sélectionner un élément HTML** identifié par son **`id`**. Une fois qu'on a sélectionner un élément HTML, on peut accéder à un de ses **attributs** HTML.   
Par exemple `document.getElementById("image2").src` permet d'accéder à l'attribut `src` (représentant le chemin vers le fichier image) de l'élément HTML d'id `"image2`


* `innerHTML` permet d'accéder au **contenu d'un élément HTML** ("ce qu'il y a entre les balises ouvrantes et fermantes")


* Pour modifier le style CSS d'un élément HTML `elmt.style.propriété = "attribut"`. Les valeurs à utiliser pour la règle de style (propriété et attribut) sont exactement les mêmes que celle utilisées dans les feuilles CSS (voir le cours sur le CSS si besoin)   
Par exemple `document.getElementById("monItem2").style.color="blue"` permet de mettre un style couleur bleue au texte de l'élément HTML d'id `"monItem2"`

### sélection d'un élément HTML

|:-:|:-:|
|`document.getElementsByClassName(nom_de_classe)`|sélectionne la **liste** des éléments ayant le nom de classe donné.|
